```python
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Time-Wasters on Social Media.csv')
def plot_distribution(data, column, kind="hist", bins=10, rotation=0):
    plt.figure(figsize=(8, 5))
    if kind == "hist":
        sns.histplot(data[column].dropna(), bins=bins, kde=True, color="skyblue")
    elif kind == "count":
        sns.countplot(x=column, data=data, palette="Set2")
        plt.xticks(rotation=rotation)
    plt.title(f"{column} Distribution")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
plot_distribution(df, "Age", kind="hist", bins=15)
```

```python
# K-Means Clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = ["Addiction Level", "Self Control", "Satisfaction", "ProductivityLoss"]

df_kmeans = df[features].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_kmeans)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

df_kmeans["Cluster"] = clusters

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette="Set1")
plt.title("KullanÄ±cÄ±larÄ±n Psikolojik Ã–zelliklerine GÃ¶re KÃ¼meleme (K-Means)")
plt.xlabel("BaÄŸÄ±mlÄ±lÄ±k (scaled)")
plt.ylabel("Ã–zdenetim (scaled)")
plt.tight_layout()
plt.show()

df_kmeans.groupby("Cluster")[features].mean().round(2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Regresyon iÃ§in kullanÄ±lacak sÃ¼tunlar
regression_features = ["Addiction Level", "Self Control", "ProductivityLoss", "KiyasEgilimi"]
target = "Satisfaction"

df_reg = df[regression_features + [target]].dropna()

X = df_reg[regression_features]
y = df_reg[target]

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_.round(3)
})

coefficients["Intercept"] = model.intercept_.round(3)
coefficients, mse, r2
```

```python
# Psikolojik ve davranÄ±ÅŸsal deÄŸiÅŸkenler arasÄ±nda korelasyon Ä±sÄ± haritasÄ± Ã§izelim
corr_cols = [
    "Addiction Level", "Self Control", "Satisfaction",
    "ProductivityLoss", "KiyasEgilimi", "Total Time Spent",
    "Engagement", "Importance Score", "Scroll Rate"
]

# GeÃ§erli sÃ¼tunlarÄ± filtrele (bazÄ±larÄ± veri setinde olmayabilir)
corr_cols_valid = [col for col in corr_cols if col in df.columns]

# Korelasyon matrisi
corr_matrix = df[corr_cols_valid].corr()

# GÃ¶rselleÅŸtirme
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("DeÄŸiÅŸkenler ArasÄ± Korelasyon IsÄ± HaritasÄ±")
plt.tight_layout()
plt.show()
```

```python
# Cinsiyet bazlÄ± tatmin ortalamalarÄ±nÄ± karÅŸÄ±laÅŸtÄ±ralÄ±m
gender_satisfaction = df.groupby("Gender")["Satisfaction"].mean().round(2)


plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Gender", y="Satisfaction", palette="pastel")
plt.title("Cinsiyete GÃ¶re Tatmin (Ruh Hali) DÃ¼zeyi DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Cinsiyet")
plt.ylabel("Tatmin Skoru")
plt.tight_layout()
plt.show()

male_satisfaction = df[df["Gender"] == "Male"]["Satisfaction"].dropna()
female_satisfaction = df[df["Gender"] == "Female"]["Satisfaction"].dropna()

from scipy.stats import ttest_ind
stat, p_val = ttest_ind(male_satisfaction, female_satisfaction, equal_var=False)

gender_satisfaction, p_val
```

```python
# Zaman kaybÄ± hissi iÃ§in "ProductivityLoss" skorunu baz alalÄ±m

time_vs_loss = df[["Total Time Spent", "ProductivityLoss"]].dropna()

from scipy.stats import pearsonr
corr, p_value = pearsonr(time_vs_loss["Total Time Spent"], time_vs_loss["ProductivityLoss"])

plt.figure(figsize=(8, 6))
sns.regplot(data=time_vs_loss, x="Total Time Spent", y="ProductivityLoss", scatter_kws={'alpha': 0.4})
plt.title("Sosyal Medyada GeÃ§irilen SÃ¼re ile Zaman KaybÄ± Hissi ArasÄ±ndaki Ä°liÅŸki")
plt.xlabel("Toplam SÃ¼re (dk)")
plt.ylabel("Zaman KaybÄ± Hissi (ProductivityLoss)")
plt.tight_layout()
plt.show()

corr, p_value
```

```python
# "Procrastination" ve "Boredom" nedenlerini kÄ±yas eÄŸilimi olarak kodlayalÄ±m
df["KiyasEgilimi"] = df["Watch Reason"].apply(lambda x: 1 if x in ["Procrastination", "Boredom"] else 0)

plt.figure(figsize=(8, 6))
sns.boxplot(x="KiyasEgilimi", y="Satisfaction", data=df, palette="Set3")
plt.title("KÄ±yas EÄŸiliminin Tatmin Ãœzerindeki Etkisi")
plt.xlabel("KÄ±yas EÄŸilimi (0=Yok, 1=Var)")
plt.ylabel("Tatmin Skoru")
plt.tight_layout()
plt.show()

from sklearn.linear_model import LinearRegression

df_reg = df[["KiyasEgilimi", "Satisfaction"]].dropna()

X = df_reg[["KiyasEgilimi"]]
y = df_reg["Satisfaction"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

coef = model.coef_[0]
intercept = model.intercept_
r2 = model.score(X, y)

coef, intercept, r2
```

```python
from scipy.stats import pearsonr, spearmanr

control_vs_addiction = df[["Self Control", "Addiction Level"]].dropna()

pearson_corr, pearson_p = pearsonr(control_vs_addiction["Self Control"], control_vs_addiction["Addiction Level"])

spearman_corr, spearman_p = spearmanr(control_vs_addiction["Self Control"], control_vs_addiction["Addiction Level"])

plt.figure(figsize=(8, 6))
sns.regplot(data=control_vs_addiction, x="Self Control", y="Addiction Level", scatter_kws={'alpha':0.5})
plt.title("Ã–zdenetim ile Sosyal Medya BaÄŸÄ±mlÄ±lÄ±ÄŸÄ± Ä°liÅŸkisi")
plt.xlabel("Ã–zdenetim (Self Control)")
plt.ylabel("BaÄŸÄ±mlÄ±lÄ±k (Addiction Level)")
plt.tight_layout()
plt.show()

pearson_corr, pearson_p, spearman_corr, spearman_p
```
